# PyTorch Inference Time

Inference time of MLP in PyTorch as a function of the number of $L$ hidden layers and the $\delta$ number of Neurons per hidden layer.

Ã“scar Amaro 31 Jan 2024

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
Llst = np.array([1,5,10,25,50])
dlst = np.array([10,25,50,100])
```

```python
timelst = np.array(pd.read_csv('inferenceTime_PyTorch.csv'))
```

```python
plt.plot(Llst, timelst)
plt.xlabel(r'L')
plt.ylabel(r'time [s]')
plt.show()
```

```python
plt.plot(dlst, timelst[0,:], label=r'L=1')
plt.plot(dlst, timelst[1,:], label=r'L=2')
plt.plot(dlst, timelst[2,:], label=r'L=3')
plt.xlabel(r'$\delta$')
plt.ylabel(r'time [s]')
plt.legend()
plt.show()
```
